### Answer 8.ii

> What does the parameter errors = ‘raise’ of method DF.drop() do? Look up the documentation and explain in your own words

> The errors parameter in the DF.drop() method refers to the behavior when errors occur during the dropping process. By default, the errors parameter is set to 'raise', which means that if any error occurs while dropping the specified columns or indices, an exception will be raised, and the operation will fail. This is useful when we want to ensure that the specified columns or indices exist in the DataFrame before attempting to drop them. If any of the specified columns or indices are missing, it prevents accidental deletion or incorrect operations.

```python
def expstandardize(x):
    return np.exp((x-x.mean())/x.std())
popDF[['y2019']].apply(expstandardize).head()
```

```python
popDF[['y2018','y2019']].apply(expstandardize).head()
```

### Answer 11.ii

```python
#compute a new column in popDF that contains the coefficient of variation for each state for the population in years 2010-2019.
popDF['coeffofvar'] = popDF[['y2010', 'y2011', 'y2012', 'y2013', 'y2014', 'y2015', 'y2016', 'y2017', 'y2018', 'y2019']].apply(
    lambda x : x.std() / x.mean(), axis = 1)

popDF[['coeffofvar']].head()
```
